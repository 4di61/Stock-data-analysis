```python
import pandas as pd

all_stock_df = pd.read_csv(
    'D:/Documents/College docs/Sem III/Mini project/data/NIFTY50_all.csv')

stock_metadata_df = pd.read_csv(
    'D:/Documents/College docs/Sem III/Mini project/data/stock_metadata.csv')
```


# DATA Cleaning


## Removing redundant columns from the dataset

```python

all_stock_df = all_stock_df.drop(columns=[
                                 "Series", "Volume", "Turnover", "Trades", "Deliverable Volume", "%Deliverble"])
all_stock_df
```

## Adding **"Industry"** column to the stock dataset

```python

for stock in stock_metadata_df.iterrows():
    all_stock_df.loc[all_stock_df["Symbol"] ==
                     stock[1].Symbol, "Industry"] = stock[1].Industry
```

## Saving dataset into a modified **.csv** file

```python

all_stock_df.to_csv(
    "D:/Documents/College docs/Sem III/Mini project/data/NIFTY50_all_modified.csv", index=False)
```
